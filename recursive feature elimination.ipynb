```python
import pandas as pd
import numpy as np
import sklearn
df = pd.read_csv('./diabetes.csv')
df.head()

```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

# memisahkan fitur dan target variabel
X = df.drop('Outcome', axis=1)
y = df['Outcome']

# membuat model Regresi Logistik
model = LogisticRegression()

# inisialisasi RFE
rfe = RFE(model, n_features_to_select=5)

# fit RFE pada data
rfe.fit(X, y)

# menampilkan hasil seleksi fitur
print('Selected Features:', X.columns[rfe.support_])
print('Ranking of Features:', rfe.ranking_)
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

# memisahkan fitur dan target variabel
X = df.drop('Outcome', axis=1)
y = df['Outcome']

# inisialisasi model Regresi Logistik dan RFE
lr = LogisticRegression()
rfe = RFE(lr, n_features_to_select=5)

# melakukan RFE pada data
rfe.fit(X, y)

# mengambil fitur yang dipilih oleh RFE
X_new = X[X.columns[rfe.support_]]

# membagi data menjadi data latih dan data uji
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_new, y, test_size=0.3, random_state=42)

# melatih model Regresi Logistik pada data latih
lr.fit(X_train, y_train)

# melakukan prediksi pada data uji
y_pred = lr.predict(X_test)

# menampilkan hasil evaluasi model
from sklearn.metrics import accuracy_score, classification_report
print('Accuracy:', accuracy_score(y_test, y_pred))
print('Classification Report:\n', classification_report(y_test, y_pred))
```

```python
#chi/univariate selection
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

# memisahkan fitur dan target variabel
X = df.drop('Outcome', axis=1)
y = df['Outcome']

# inisialisasi selector
selector = SelectKBest(chi2, k=5)

# fit selector pada data
X_new = selector.fit_transform(X, y)

# menampilkan hasil seleksi fitur
print('Selected Features:', X.columns[selector.get_support()])
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

# memisahkan fitur dan target variabel
X = df2.drop('diagnosis', axis=1)
y = df2['diagnosis']

# membuat model Regresi Logistik
model = LogisticRegression()

# menghitung jumlah fitur yang akan dipilih (70% dari total fitur)
num_features = int(round(len(X.columns) * 0.7))

# inisialisasi RFE dengan jumlah fitur yang telah dihitung sebelumnya
rfe = RFE(model, n_features_to_select=num_features)

# fit RFE pada data
rfe.fit(X, y)

# menampilkan hasil seleksi fitur
print('Selected Features:', X.columns[rfe.support_])
print('Ranking of Features:', rfe.ranking_)
```

```python
#dengan knn
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score


# memisahkan fitur dan target variabel
a = df2[['radius_mean', 'texture_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'texture_se', 'perimeter_se',
       'area_se', 'compactness_se', 'concavity_se', 'radius_worst',
       'texture_worst', 'perimeter_worst', 'area_worst', 'smoothness_worst',
       'compactness_worst', 'concavity_worst', 'concave points_worst',
       'symmetry_worst', 'fractal_dimension_worst']]
b = df2['diagnosis']

# membagi data menjadi data latih dan data uji
X_train, X_test, y_train, y_test = train_test_split(a, b, test_size=0.2, random_state=42)

# inisialisasi model KNN
knn = KNeighborsClassifier(n_neighbors=5)

# fit model pada data latih
knn.fit(X_train, y_train)

# membuat prediksi pada data uji
y_pred = knn.predict(X_test)

# menghitung akurasi model
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, pos_label='M') # atau pos_label='B'
recall = recall_score(y_test, y_pred,pos_label='M')
f1 = f1_score(y_test, y_pred,pos_label='M')

# menampilkan hasil akurasi, presisi, recall, dan F1 score
print('Accuracy:', accuracy)
print('Precision:', precision)
print('Recall:', recall)
print('F1 Score:', f1)

```
